        B0015 - Na semana passada teve dor de cabeça?

        B00112 - Na semana passada teve dor muscular?

        B0031 - Providência tomada para recuperar dos sintomas foi ficar em casa

        B0041 - Local que buscou atendimento foi posto de saúde/Unidade básica de saúde /Equipe de Saúde da Família (médico, enfermeiro, técnico de enfermagem ou agente comunitário de saúde)

    Parte Suplementar 01 - Características da habitação

```python
# Renomear colunas
df = df.rename(columns = {'V1012': 'Semana',
                          'V1013': 'Mes',
                          'V1022': 'Zona',
                          'V1030': 'Populacao',
                          'A002': 'Idade',
                          'A003': 'Sexo',
                          'A004': 'Cor',
                          'B0011': 'Febre',
                          'B0012': 'Tosse',
                          'B0013': 'Dor_de_garganta',
                          'B0014': 'Dificuldade_para_respirar',
                          'B0015': 'Dor_de_cabeca',
                          'B00111': 'Perda_de_cheiro_ou_sabor',
                          'B00112': 'Dor_muscular',
                          'B002': 'Foi_a_algum_estabelecimento_de_saude',
                          'B0031': 'Apenas_ficou_em_casa',
                          'B0034': 'Comprou_e_ou_tomou_remedio_por_orientacao_medica',
                          'B0041': 'Foi_em_Posto_ou_UBS_ou_ESF',
                          'B0042': 'Foi_em_pronto_socorro_do_SUS_ou_UPA',
                          'B0043': 'Foi_em_hospital_do_SUS',
                          'B005': 'Internado_por_um_dia_ou_mais',
                          'B006': 'Sedado_e_entubado_e_colocado_em_respiracao_artificial_com_ventilador',
                          'B007': 'Plano_de_saude',
                          'C001': 'Trabalhou_ou_fez_algum_bico_por_pelo_menos_uma_hora_na_semana_passada',
                          'C007': 'Tipo_de_vinculo_trabalhista',
                          'C007B': 'CTPS_ou_funcionario_publico_estatutario',
                          'C01012': 'Valor_recebido_em_dinheiro',
                          'C013': 'Trabalho_remoto',
                          'C014': 'INSS',
                          'D0053': 'Auxilios_emergencial_do_coronavirus',
                          'F001': 'Tipo_do_domicilio'
                        })


# Criar coluna Mes/Semana
def padronizar_mes_semana(row):
    return str(int(row["Mes"])) + "/" + str(int(row["Semana"]))
df["Mes_Semana"] = df.apply(padronizar_mes_semana, axis = 1)


# Padronizar UF
def padronizar_uf(uf):
    dict_ufs = {
                11: 'RO',
                12: 'AC',
                13: 'AM',
                14: 'RR',
                15: 'PA',
                16: 'AP',
                17: 'TO',
                21: 'MA',
                22: 'PI',
                23: 'CE',
                24: 'RN',
                25: 'PB',
                26: 'PE',
                27: 'AL',
                28: 'SE',
                29: 'BA',
                31: 'MG',
                32: 'ES',
                33: 'RJ',
                35: 'SP',
                41: 'PR',
                42: 'SC',
                43: 'RS',
                50: 'MS',
                51: 'MT',
                52: 'GO',
                53: 'DF'
                }
    return dict_ufs[uf]
df['UF'] = df['UF'].apply(padronizar_uf)


# Padronização da zona do domicílio
def padronizar_zona(zona):
    if zona == 1:
        return 'Urbana'
    elif zona == 2:
        return 'Rural'
df['Zona'] = df['Zona'].apply(padronizar_zona)


# Padronização do sexo
def padronizar_sexo(sexo):
    if sexo == 1:
        return 'Homem'
    elif sexo == 2:
        return 'Mulher'
df['Sexo'] = df['Sexo'].apply(padronizar_sexo)


# Padronização da cor
def padronizar_cor(cor):
    if cor == 1:
        return 'Branca'
    elif cor == 2:
        return 'Preta'
    elif cor == 3:
        return 'Amarela'
    elif cor == 4:
        return 'Parda'
    elif cor == 5:
        return 'Indígena'
    elif cor == 9:
        return 'Ignorado'
df['Cor'] = df['Cor'].apply(padronizar_cor)


# Padronizar sintomas
def padronizar_sintoma(sintoma):
    if sintoma == 1:
        return 'Sim'
    elif sintoma == 2:
        return 'Não'
    elif sintoma == 3:
        return 'Não sabe'
    elif sintoma == 9:
        return 'Ignorado'
df['Febre']                     = df['Febre'].apply(padronizar_sintoma)
df['Tosse']                     = df['Tosse'].apply(padronizar_sintoma)
df['Dor_de_garganta']           = df['Dor_de_garganta'].apply(padronizar_sintoma)
df['Dificuldade_para_respirar'] = df['Dificuldade_para_respirar'].apply(padronizar_sintoma)
df['Dor_de_cabeca']             = df['Dor_de_cabeca'].apply(padronizar_sintoma)
df['Perda_de_cheiro_ou_sabor']  = df['Perda_de_cheiro_ou_sabor'].apply(padronizar_sintoma)
df['Dor_muscular']              = df['Dor_muscular'].apply(padronizar_sintoma)


# Padronizar ação que o entrevistado tomou
def padronizar_acao(acao):
    if acao == 1:
        return 'Sim'
    elif acao == 2:
        return 'Não'
    elif acao == 9:
        return 'Ignorado'
    elif acao == '':
        return 'Não aplicável'
df['Foi_a_algum_estabelecimento_de_saude'] = df['Foi_a_algum_estabelecimento_de_saude'].apply(padronizar_acao)
df['Apenas_ficou_em_casa'] = df['Apenas_ficou_em_casa'].apply(padronizar_acao)
df['Comprou_e_ou_tomou_remedio_por_orientacao_medica'] = df['Comprou_e_ou_tomou_remedio_por_orientacao_medica'].apply(padronizar_acao)
df['Foi_em_Posto_ou_UBS_ou_ESF'] = df['Foi_em_Posto_ou_UBS_ou_ESF'].apply(padronizar_acao)
df['Foi_em_pronto_socorro_do_SUS_ou_UPA'] = df['Foi_em_pronto_socorro_do_SUS_ou_UPA'].apply(padronizar_acao)
df['Foi_em_hospital_do_SUS'] = df['Foi_em_hospital_do_SUS'].apply(padronizar_acao)
df['Internado_por_um_dia_ou_mais'] = df['Internado_por_um_dia_ou_mais'].apply(padronizar_acao)
df['Sedado_e_entubado_e_colocado_em_respiracao_artificial_com_ventilador'] = df['Sedado_e_entubado_e_colocado_em_respiracao_artificial_com_ventilador'].apply(padronizar_acao)
df['Plano_de_saude'] = df['Plano_de_saude'].apply(padronizar_acao)


# Padronizar o tipo de vínculo trabalhista
def padronizar_tipo_trabalho(tipo):
    if tipo == 1:
        return 'Trabalhador doméstico (empregado doméstico, cuidados, babá)'
    elif tipo == 2:
        return 'Militar do exercito, marinha ou aeronáutica'
    elif tipo == 3:
        return 'Policial militar ou bombeiro militar'
    elif tipo == 4:
        return 'Empregado do setor privado'
    elif tipo == 5:
        return 'Empregado do setor público (inclusive empresas de economia mista)'
    elif tipo == 6:
        return 'Empregador'
    elif tipo == 7:
        return 'Conta própria'
    elif tipo == 8:
        return 'Trabalhador familiar não remunerado em ajuda a membro do domicílio ou parente'
    elif tipo == 9:
        return 'Estava fora do mercado de trabalho (fazia apenas afazeres domésticos, cuidados de pessoas ou produção para próprio consumo)'
    elif tipo == '':
        return 'Não aplicável'
df['Tipo_de_vinculo_trabalhista'] = df['Tipo_de_vinculo_trabalhista'].apply(padronizar_tipo_trabalho)


#  Padronizar o fato de ter trabalhado ou feito algum bico por pelo menos uma hora na semana passada
def padronizar_trabalho(tipo):
    if tipo == 1:
        return 'Sim'
    elif tipo == 2:
        return 'Não'
    elif tipo == '':
        return 'Não aplicável'
df['Trabalhou_ou_fez_algum_bico_por_pelo_menos_uma_hora_na_semana_passada'] = df['Trabalhou_ou_fez_algum_bico_por_pelo_menos_uma_hora_na_semana_passada'].apply(padronizar_trabalho)


# Padronizar a questão de ter CTPS ou não
def padronizar_ctps(tipo):
    if tipo == 1:
        return 'Sim, tem carteira de trabalho assinada'
    elif tipo == 2:
        return 'Sim, é servidor público estatutário'
    elif tipo == 3:
        return 'Não'
    elif tipo == '':
        return 'Não aplicável'
df['CTPS_ou_funcionario_publico_estatutario'] = df['CTPS_ou_funcionario_publico_estatutario'].apply(padronizar_ctps)


# Padronizar se trabalha remotamente
def padronizar_trabalho_remoto(tipo):
    if tipo == 1:
        return 'Sim'
    elif tipo == 2:
        return 'Não'
    elif tipo == '':
        return 'Não aplicável'
df['Trabalho_remoto'] = df['Trabalho_remoto'].apply(padronizar_trabalho_remoto)


# Padronizar se contribui para o INSS
def padronizar_inss(tipo):
    if tipo == 1:
        return 'Sim'
    elif tipo == 2:
        return 'Não'
    elif tipo == '':
        return 'Não aplicável'
df['INSS'] = df['INSS'].apply(padronizar_trabalho_remoto)


# Padronizar ação que o entrevistado tomou
def padronizar_tipo_domicilio(tipo):
    if tipo == 1:
        return 'Próprio (já pago)'
    elif tipo == 2:
        return 'Próprio (ainda pagando)'
    elif tipo == 3:
        return 'Alugado'
    elif tipo == 4:
        return 'Cedido por empregador'
    elif tipo == 5:
        return 'Cedido por familiar'
    elif tipo == 6:
        return 'Cedido de outra forma'
    elif tipo == 7:
        return 'Outra condição'
df['Tipo_do_domicilio'] = df['Tipo_do_domicilio'].apply(padronizar_tipo_domicilio)
```
